```python
import pandas as pd
import os

# File paths
file_path_1 = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\abetaUniprot.csv'
file_path_2 = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\Abeta_analysis.csv'
output_file_path = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\new_merged_data_cleaned.csv'  # Update this path as needed

# Read the CSV files
df1 = pd.read_csv(file_path_1)
df2 = pd.read_csv(file_path_2)

# Merge the DataFrames
merged_df = pd.merge(df1, df2, on='Protein_Name', how='outer')

# Combine the columns with suffixes
for col in df1.columns:
    if col != 'Protein_Name' and col in df2.columns:
        merged_df[col] = merged_df[col + '_x'].combine_first(merged_df[col + '_y'])
        merged_df.drop(columns=[col + '_x', col + '_y'], inplace=True)

# Drop duplicate rows
merged_df = merged_df.drop_duplicates().reset_index(drop=True)

# Save the cleaned merged DataFrame
merged_df.to_csv(output_file_path, index=False)

print(f"Data merged, duplicates removed, and saved successfully to {output_file_path}!")
```

```python
import pandas as pd
import gzip
import shutil
import os

# Read the existing data
existing_data_path = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\new_merged_data_cleaned.csv'
existing_data = pd.read_csv(existing_data_path)

print("Existing data shape:", existing_data.shape)
print("\nFirst few rows of existing data:")
print(existing_data.head())

# Decompress the gzipped Excel file
compressed_file = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\uniprotkb_APP_AND_model_organism_9606_2024_07_30.xlsx.gz'
decompressed_file = compressed_file[:-3]  # Remove the .gz extension

with gzip.open(compressed_file, 'rb') as f_in:
    with open(decompressed_file, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

# Read the UniProt data
uniprot_data = pd.read_excel(decompressed_file)

# Remove the decompressed file to save space
os.remove(decompressed_file)

print("\nUniProt data shape:", uniprot_data.shape)
print("\nFirst few rows of UniProt data:")
print(uniprot_data.head())

# Merge the datasets
# We'll use an outer join to keep all data from both datasets
# You may need to adjust the 'on' parameter depending on which column(s) you want to use for merging
merged_data = pd.merge(existing_data, uniprot_data, how='outer', left_on='Protein_ID', right_on='Entry')

# Display info about the merged dataset
print("\nMerged data shape:", merged_data.shape)

# Check for any duplicate columns and remove them if necessary
merged_data = merged_data.loc[:,~merged_data.columns.duplicated()]

# Save the merged data to a new Excel file
output_path = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\merged_alzheimers_protein_data.xlsx'
merged_data.to_excel(output_path, index=False)

print(f"\nMerged data saved to {output_path}")

# Display the first few rows of the merged data
print("\nFirst few rows of merged data:")
print(merged_data.head())
```

```python
import pandas as pd

# Corrected file path with file name
file_path = r'C:\Users\14807\OneDrive\Spreadsheets\Data Science Capstone\AI_ML_Biotech\Data Collection\cleaned_alzheimers_data.xlsx'

# Load the data
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Inspect columns with high number of missing values
missing_values = df.isnull().sum()
high_missing_columns = missing_values[missing_values > 500]
print(high_missing_columns)

# Decide on a strategy: Drop columns or impute values
df_cleaned = df.drop(columns=high_missing_columns.index)

# Filling numerical columns with the mean
numeric_columns = df_cleaned.select_dtypes(include=['number']).columns
df_cleaned[numeric_columns] = df_cleaned[numeric_columns].fillna(df_cleaned[numeric_columns].mean())

# Filling categorical columns with the mode
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
for column in categorical_columns:
    df_cleaned[column].fillna(df_cleaned[column].mode()[0], inplace=True)

# Save the cleaned data to a new file
df_cleaned.to_excel('further_cleaned_alzheimers_data.xlsx', index=False)
```

```python
import requests

# Set the base URL for ProteomicsDB API
base_url = "https://www.proteomicsdb.org/proteomicsdb/logic/api/"

# Specify the UniProt identifier for the APP gene
protein_filter = 'P05067'

# Define the endpoint and parameters
endpoint = f"proteinpeptideresult.xsodata/InputParams(PROTEINFILTER='{protein_filter}')/Results"
params = {
    "$select": "ENTRY_NAME,PROTEIN_NAME,UNIQUE_IDENTIFIER,TAXCODE,CHROMOSOME_NAME,GENE_NAME,STRAND,PEPTIDE_SEQUENCE,PEPTIDE_MASS,START_POSITION,END_POSITION,PEPTIDE_SCORE,PROTEIN_Q_VALUE,PEPTIDE_Q_VALUE,SEARCH_ENGINE,ISUNIQUE,ISUNIQUE_PROTEIN,PROJECT_NAME,PROJECT_DESCRIPTION,EXPERIMENT_NAME,EXPERIMENT_ID,EXPERIMENT_DESCRIPTION,PUBMEDID",
    "$format": "json"
}

# Send the request to the API
response = requests.get(base_url + endpoint, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    # Process the data as needed
    print(data)
else:
    print(f"Error: {response.status_code}, {response.text}")
```

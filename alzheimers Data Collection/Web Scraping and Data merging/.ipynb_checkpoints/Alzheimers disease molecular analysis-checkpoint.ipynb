```python
import requests
import pandas as pd
import io

# URL for UniProt API
url = "https://rest.uniprot.org/uniprotkb/stream"

# Parameters for the API request
params = {
    "query": "(protein_name:amyloid beta 40 42) AND (reviewed:true)",
    "format": "tsv",
    "fields": "accession,id,protein_name,sequence,length,mass,cc_disease,cc_function,cc_subcellular_location"
}

# Make the API request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Read the TSV data into a pandas DataFrame
    df = pd.read_csv(io.StringIO(response.text), sep='\t')
    
    # Process the data
    processed_data = []
    for _, row in df.iterrows():
        processed_data.append({
            'Protein_Name': row.get('Protein names', 'N/A'),
            'Protein_ID': f"UniProt:{row.get('Entry', 'N/A')}",
            'Sequence': row.get('Sequence', 'N/A'),
            'Structure_File': 'N/A',
            'Disease_Related': 'Yes' if pd.notna(row.get('Disease')) else 'No',
            'Molecular_Weight': row.get('Mass', 'N/A'),
            'Isoelectric_Point': 'N/A',
            'Source': f"UniProt:{row.get('Entry', 'N/A')}",
            'Secondary_Structure_Alpha': 'N/A',
            'Secondary_Structure_Beta': 'N/A',
            'Secondary_Structure_Others': 'N/A',
            'Radius_of_Gyration': 'N/A',
            'Total_SASA': 'N/A',
            'End_to_End_Distance': 'N/A',
            'RMSD_to_Abeta40': 'N/A',
            'Notes': row.get('Function', 'N/A')
        })

    # Create final dataframe
    final_df = pd.DataFrame(processed_data)

    # Display the first few rows
    display(final_df.head(10))

    # Save to CSV
    final_df.to_csv('amyloid_beta_data.csv', index=False)
    print("Data saved to 'amyloid_beta_data.csv'")
else:
    print(f"Error: {response.status_code}")
    print(response.text)

# Print out the column names we actually received
print("Columns received from UniProt:")
print(df.columns.tolist())
```

```python
import requests
import pandas as pd
import io
from Bio.SeqUtils.ProtParam import ProteinAnalysis

def get_uniprot_data(uniprot_id):
    url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}"
    response = requests.get(url, params={"format": "json"})
    if response.status_code == 200:
        data = response.json()
        sequence = data['sequence']['value']
        function = data.get('comments', [{}])[0].get('text', [''])[0] if data.get('comments') else ''
        return sequence, function
    return None, None

def calculate_protein_properties(sequence):
    if not sequence:
        return {
            'Molecular_Weight': 'N/A',
            'Isoelectric_Point': 'N/A',
            'Aromaticity': 'N/A',
            'Instability_Index': 'N/A',
            'Gravy': 'N/A',
        }
    analysis = ProteinAnalysis(sequence)
    return {
        'Molecular_Weight': analysis.molecular_weight(),
        'Isoelectric_Point': analysis.isoelectric_point(),
        'Aromaticity': analysis.aromaticity(),
        'Instability_Index': analysis.instability_index(),
        'Gravy': analysis.gravy(),
    }

# Define amyloid beta related proteins to analyze (you can add more UniProt IDs here)
uniprot_ids = [
    'P05067', 'P51693', 'Q06481', 'O00213', 'Q92870', 'Q92624', 'B4DMD5', 
    'B4DGD0', 'A0A140VJC8', 'B4DQM1', 'A0A218KGR2', 'E9PG40', 'Q99767', 
    'O96018', 'Q02410', 'B4DJT9', 'B4DM00', 'A0A0A0MRG2', 'Q7Z5R6', 'A4D1B5', 
    'B7Z313', 'A0A8F9R5D5', 'H7C0V9', 'L7XCZ9', 'Q9BYY9', 'L7XE61', 'Q9UED0', 
    'B4E3I5', 'B4DTV1', 'B4DKY3', 'E9PPD0', 'B2R5S2', 'F5GZ08', 'A0A0A0MRG2', 
    'B7Z4Q2', 'Q5I0G1'
]
data = []

for uniprot_id in uniprot_ids:
    sequence, function = get_uniprot_data(uniprot_id)
    if sequence:
        protein_props = calculate_protein_properties(sequence)
        
        entry = {
            'Protein_Name': f"APP (UniProt:{uniprot_id})",
            'Protein_ID': uniprot_id,
            'Sequence': sequence,
            'Disease_Related': 'Yes',
            'Molecular_Weight': protein_props['Molecular_Weight'],
            'Isoelectric_Point': protein_props['Isoelectric_Point'],
            'Aromaticity': protein_props['Aromaticity'],
            'Instability_Index': protein_props['Instability_Index'],
            'Gravy': protein_props['Gravy'],
            'Source': f"UniProt:{uniprot_id}",
            'Notes': f'Full APP protein. Function: {function}'
        }
        data.append(entry)
        
        # Add Abeta40 and Abeta42 fragments
        abeta_fragments = {
            'Abeta40': (672, 711),
            'Abeta42': (672, 713)
        }
        for fragment_name, (start, end) in abeta_fragments.items():
            if start <= len(sequence) and end <= len(sequence):
                fragment_sequence = sequence[start-1:end]
                fragment_props = calculate_protein_properties(fragment_sequence)
                
                fragment_entry = {
                    'Protein_Name': fragment_name,
                    'Protein_ID': f"Fragment of UniProt:{uniprot_id}",
                    'Sequence': fragment_sequence,
                    'Disease_Related': 'Yes',
                    'Molecular_Weight': fragment_props['Molecular_Weight'],
                    'Isoelectric_Point': fragment_props['Isoelectric_Point'],
                    'Aromaticity': fragment_props['Aromaticity'],
                    'Instability_Index': fragment_props['Instability_Index'],
                    'Gravy': fragment_props['Gravy'],
                    'Source': f"Derived from UniProt:{uniprot_id}",
                    'Notes': f'Fragment of APP, positions {start}-{end}'
                }
                data.append(fragment_entry)
            else:
                print(f"Warning: Unable to extract {fragment_name} (positions {start}-{end}) from sequence of length {len(sequence)}")
    else:
        print(f"Warning: Unable to retrieve data for UniProt ID {uniprot_id}")

if data:
    df = pd.DataFrame(data)
    display(df)
    df.to_csv('amyloid_beta_data.csv', index=False)
    print("Data saved to 'amyloid_beta_data.csv'")
else:
    print("No data was retrieved. Check your UniProt IDs and internet connection.")
```
